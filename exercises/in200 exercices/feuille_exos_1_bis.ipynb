```python
import tkinter as tk 

graphique = tk.Tk()
graphique.title('exercice 5')

x_rouge = 200
x_bleue = 400
couleur = ["orange red","DeepSkyBlue2"]

def bouger_ligne(event):
    global x_rouge, x_bleue, couleur
    x, y = event.x, event.y 
    canevas.itemconfigure(ligne_bleue, fill=couleur[0])
    canevas.itemconfigure(ligne_rouge, fill=couleur[1])
    couleur[0], couleur[1] = couleur[1], couleur[0]
    if (x <= x_rouge) and (x <= x_bleue) :
        x_rouge -= 10
        x_bleue -= 10
        canevas.move(ligne_bleue, -10, 0)
        canevas.move(ligne_rouge, -10, 0)
    elif (x >= x_rouge) and (x >= x_bleue) :
        x_rouge += 10
        x_bleue += 10
        canevas.move(ligne_bleue, +10, 0)
        canevas.move(ligne_rouge, +10, 0)
    elif (x >= x_rouge) and (x <= x_bleue) :
        x_rouge += 10
        x_bleue -= 10
        canevas.move(ligne_bleue, -10, 0)
        canevas.move(ligne_rouge, +10, 0)

def again():
    global x_rouge, x_bleue, couleur
    couleur = ["orange red","DeepSkyBlue2"]
    canevas.itemconfigure(ligne_bleue, fill=couleur[1])
    canevas.itemconfigure(ligne_rouge, fill=couleur[0])
    canevas.move(ligne_rouge, 200-x_rouge, 0)
    canevas.move(ligne_bleue, 400-x_bleue, 0)
    x_rouge = 200
    x_bleue = 400
    
canevas = tk.Canvas(graphique, width=600, heigh=600, bg='white')
bouton_redemarrer = tk.Button(graphique, text='redémarrer', bg='VioletRed1', width=28, command=again)

canevas.grid(column=0, row=0, columnspan=3)
bouton_redemarrer.grid(column=1, row=1)

ligne_rouge = canevas.create_line(200, 0, 200, 600, fill="orange red", width=3)
ligne_bleue = canevas.create_line(400, 0, 400, 600, fill="DeepSkyBlue2", width=3)
canevas.bind('<Button-1>', bouger_ligne)

graphique.mainloop()

```
